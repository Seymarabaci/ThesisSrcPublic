```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
# Load the red band and NIR band rasters
with rasterio.open(r'/yeni_sentinel2/B8_Soke11.tif') as nir_ras:
    nir = nir_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)

# Calculate NDVI
NDVI = (nir-red) / (nir+red)

NDVI_filtered = np.where(NDVI<0.000001,np.nan,NDVI)


# Create a figure and plot the NDVI image using the "RdYlGn" color map
fig, ax = plt.subplots()
im = ax.imshow(NDVI_filtered, cmap="RdYlGn")

# Add a color bar to the figure
cbar = fig.colorbar(im, ax=ax)

# Show the figure
plt.show()
from rasterio.plot import show_hist
show_hist(source=NDVI_filtered, bins=50, title='Histogram of NDVI 1_1', 
          histtype='stepfilled', alpha=0.5)
```
