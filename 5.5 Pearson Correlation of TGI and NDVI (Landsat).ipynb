```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Load the red band and NIR band rasters
with rasterio.open(r'/Landsat9/L9_B5_Soke11.tif') as nir_ras:
    nir = nir_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)

# Calculate NDVI
NDVI = (nir - red) / (nir + red)

with rasterio.open(r'/Landsat9/L9_B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B3_Soke11.tif') as green_ras:
    green = green_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B2_Soke11.tif') as blue_ras:
    blue = blue_ras.read(1)

# Calculate TGI
TGI = green - 0.39 * red - 0.61 * blue
TGI_filtered = np.where(TGI < 0.000001, np.nan, TGI)

values1 = TGI_filtered
values2 = NDVI

# Remove NaN values from the arrays
valid_indices = np.logical_not(np.logical_or(np.isnan(values1), np.isnan(values2)))
values1 = values1[valid_indices]
values2 = values2[valid_indices]

# Calculate the Pearson correlation between the two arrays
correlation = np.corrcoef(values1, values2)[0, 1]

# Print the Pearson correlation
print("correlation:", correlation)

# Create a scatter plot
plt.scatter(values1.flatten(), values2.flatten())
plt.xlabel("TGI Values")
plt.ylabel("NDVI Values")
plt.title("Scatter Plot of TGI and NDVI 1_1")

# Add the Pearson correlation line
z = np.polyfit(values1.flatten(), values2.flatten(), 1)
p = np.poly1d(z)
plt.plot(values1.flatten(), p(values1.flatten()), color='red')

# Add the Pearson correlation line
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')
# Show the plot

plt.show()
```
