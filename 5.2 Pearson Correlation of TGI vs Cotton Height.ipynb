```python
import rasterio
import numpy as np
from scipy.stats import pearsonr
from skimage.transform import resize
import matplotlib.pyplot as plt

# Open the raster dataset for cotton height
with rasterio.open('/Resampled/repro_DEM_11_mask.tif') as src:
    # Read the raster values into a numpy array
    raster_values = src.read(1)

# Open the raster datasets for TGI
with rasterio.open('/TGI_uav/Soke11_TGI.tif') as src:
    tgi = src.read(1)

# Resize the arrays to match their dimensions
resized_raster_values = resize(raster_values, tgi.shape, mode='constant')

# Preprocess the arrays to handle NaN or Inf values
resized_raster_values = np.nan_to_num(resized_raster_values)
tgi = np.nan_to_num(tgi)

# Calculate Pearson correlation
correlation, p_value = pearsonr(resized_raster_values.flatten(), tgi.flatten())

print(f"Pearson correlation: {correlation}")
print(f"P-value: {p_value}")

# Filter data for cotton height > 10.0
filtered_raster_values = resized_raster_values[resized_raster_values > 12.0]
filtered_tgi = tgi[resized_raster_values > 12.0]

# Create scatter plot
plt.scatter(filtered_raster_values.flatten(), filtered_tgi.flatten())
plt.xlabel('Cotton Height')
plt.ylabel('TGI')
plt.title('Scatter Plot of Cotton Height vs TGI 1_1')

# Add the Pearson correlation line
z = np.polyfit(filtered_raster_values.flatten(), filtered_tgi.flatten() , 1)
p = np.poly1d(z)
plt.plot(filtered_raster_values.flatten(), p(filtered_raster_values.flatten()), color='red')

# Add the Pearson correlation annotation
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')

# Show the plot
plt.show()
```
