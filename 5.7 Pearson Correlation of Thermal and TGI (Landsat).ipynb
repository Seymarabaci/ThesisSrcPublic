```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

with rasterio.open(r'/Landsat9/L9_B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B3_Soke11.tif') as green_ras:
    green = green_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B2_Soke11.tif') as blue_ras:
    blue = blue_ras.read(1)

# Calculate TGI
TGI = green - 0.39 * red - 0.61 * blue
TGI_filtered = np.where(TGI < 0.000001, np.nan, TGI)

# Open the raster dataset
with rasterio.open(r'/Landsat9_thermal/Soke1_1_T.tif') as src:
    # Read the raster values into a numpy array
    raster_values = src.read(1)

# Extract the plant heights from the raster values
thermal = raster_values[raster_values > 0]

# Resize the NDVI array to have the same dimensions as thermal
TGI_filtered = np.resize(TGI_filtered, thermal.shape)

# Remove NaN values from the arrays
valid_indices = np.logical_not(np.logical_or(np.isnan(thermal), np.isnan(TGI_filtered)))
thermal = thermal[valid_indices]
TGI_filtered = TGI_filtered[valid_indices]

# Calculate the Pearson correlation coefficient
correlation, _ = pearsonr(thermal, TGI_filtered)

# Print the Pearson correlation
print("Correlation:", correlation)

# Create a scatter plot
plt.scatter(thermal, TGI_filtered)
plt.xlabel("Thermal of Landsat")
plt.ylabel("TGI Values")
plt.title("Scatter Plot of Thermal Band and TGI 1_1")

# Add the Pearson correlation line
z = np.polyfit(thermal, TGI_filtered, 1)
p = np.poly1d(z)
plt.plot(thermal, p(thermal), color='red')

# Add the Pearson correlation line
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')

# Show the plot
plt.show()
```
