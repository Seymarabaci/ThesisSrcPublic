```python
import rasterio
import matplotlib.pyplot as plt

# Load the red band and NIR band rasters
with rasterio.open(r'/Landsat9/L9_B5_Soke11.tif') as nir_ras:
    nir = nir_ras.read(1)
with rasterio.open(r'/Landsat9/L9_B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)

# Calculate NDVI
NDVI = (nir-red) / (nir+red)

# Create a figure and plot the NDVI image using the "RdYlGn" color map
fig, ax = plt.subplots()
im = ax.imshow(NDVI, cmap="RdYlGn")

# Add a color bar to the figure
cbar = fig.colorbar(im, ax=ax)

# Show the figure
plt.show()
from rasterio.plot import show_hist
show_hist(source=NDVI, bins=50, title='Histogram of NDVI 1_1',
          histtype='stepfilled', alpha=1)
```
