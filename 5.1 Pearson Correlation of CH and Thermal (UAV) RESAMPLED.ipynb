```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Open the raster datasets
with rasterio.open(r'/Resampled/repro_thermal_11.tif') as src1:
    # Read the raster values into a numpy array
    raster_values1 = src1.read(1)

with rasterio.open(r'/Resampled/repro_DEM_11.tif') as src2:
    # Read the raster values into a numpy array
    raster_values2 = src2.read(1)

# Extract the plant heights from the raster values
thermal = raster_values1[raster_values1 > 0]
cotton_height = raster_values2[raster_values2 > 0]

# Remove NaN values from the arrays
valid_indices = np.logical_and(thermal > 0, cotton_height > 0)
thermal = thermal[valid_indices]
cotton_height = cotton_height[valid_indices]

# Convert arrays to float data type
thermal = thermal.astype(float)
cotton_height = cotton_height.astype(float)

# Calculate the Pearson correlation coefficient
correlation, _ = pearsonr(thermal, cotton_height)

# Print the Pearson correlation
print("Correlation:", correlation)

# Create a scatter plot
plt.scatter(thermal, cotton_height)
plt.xlabel("Thermal values from UAV")
plt.ylabel("Cotton Heights from DEM")
plt.title("Scatter Plot of Thermal and Cotton Height 1_1")

# Add the Pearson correlation line
z = np.polyfit(thermal, cotton_height, 1)
p = np.poly1d(z)
plt.plot(thermal, p(thermal), color='red')

# Add the Pearson correlation annotation
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')

# Show the plot
plt.show()
```
