```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Open the raster dataset
with rasterio.open(r'/Orthomosaics/Soke1_1.tif') as src:
    # Read the raster values into a numpy array
    raster_values = src.read(1)
    
# Extract the plant heights from the raster values
plant_heights = raster_values[raster_values > 0]


# Create a histogram
hist_range = (0,45)
colors = ['red']
plt.hist(plant_heights, bins = 50, color=colors, range=hist_range)

# Add a title and labels
plt.title("Histogram of UAV Thermal 1_1")
plt.xlabel("DN")
plt.ylabel("Frequency")



# Show the plot
plt.show()
```
