```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

# Load the red band and NIR band rasters
with rasterio.open(r'/yeni_sentinel2/B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B3_Soke11.tif') as green_ras:
    green = green_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B2_Soke11.tif') as blue_ras:
    blue = blue_ras.read(1)

# Calculate TGI
TGI = -0.5 * ((red - green) * 175 - (red - blue) * 105)

# Load the red band and NIR band rasters
with rasterio.open(r'/yeni_sentinel2/B8_Soke11.tif') as nir_ras:
    nir = nir_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)

# Calculate NDVI
NDVI = (nir - red) / (nir + red)

values1 = TGI
values2 = NDVI

# Check the dimensions and shapes of the arrays
print('values1 shape:', values1.shape)
print('values2 shape:', values2.shape)

# Resize the smaller array to have the same shape as the larger array along the second axis
if values1.shape[1] > values2.shape[1]:
    values2 = np.resize(values2, (values2.shape[0], values1.shape[1]))
else:
    values1 = np.resize(values1, (values1.shape[0], values2.shape[1]))

# Concatenate the two arrays
values3 = np.concatenate((values1, values2))

# Calculate the Pearson correlation between the two arrays
correlation = np.corrcoef(values3)[0, 1]

# Print the Pearson correlation
print("correlation:", correlation)

# Create a scatter plot
plt.scatter(values1.flatten(), values2.flatten())
plt.xlabel("TGI Values")
plt.ylabel("NDVI Values")
plt.title("Scatter Plot of TGI and NDVI 1_1")

# Add the Pearson correlation line
z = np.polyfit(values1.flatten(), values2.flatten(), 1)
p = np.poly1d(z)
plt.plot(values1.flatten(), p(values1.flatten()), color='red')

# Add the Pearson correlation line
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')
# Show the plot

plt.show()
```
