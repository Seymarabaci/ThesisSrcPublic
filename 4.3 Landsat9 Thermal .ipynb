```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Open the raster dataset
with rasterio.open(r'/Landsat9_thermal/Soke1_1_T.tif') as src:
    # Read the raster values into a numpy array
    raster_values = src.read(1)
    
# Extract the plant heights from the raster values
plant_heights = raster_values[raster_values > 0]

# Create a histogram
hist_range = (45000,47250)
colors = ['orange']
plt.hist(plant_heights, bins = 50, range=hist_range, color=colors)


# Add a title and labels
plt.title("Histogram of Landsat9 Thermal Band 1_1")
plt.xlabel("Frequency")
plt.ylabel("DN")

# Show the plot
plt.show()
```
