```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Open the raster datasets
with rasterio.open(r'/Resampled/repro_DEM_11.tif') as src2:
    # Read the raster values into a numpy array
    raster_values2 = src2.read(1)

# Extract the plant heights from the raster values
cotton_height = raster_values2[raster_values2 > 0]


with rasterio.open(r'/Resampled/repro_wide_11.tif') as red_ras:
    red = red_ras.read(1)
with rasterio.open(r'/Resampled/repro_wide_11.tif') as green_ras:
    green = green_ras.read(2)
with rasterio.open(r'/Resampled/repro_wide_11.tif') as blue_ras:
    blue = blue_ras.read(3)
TGI = (green - (blue + red) / 2) / np.sqrt(green + (blue + red) / 2)
#TGI = TGI_before[TGI_before > 7.0]

# Resize arrays to match each other's shape
TGI_resized = np.resize(TGI, cotton_height.shape)

# Flatten the arrays to 1D
values1 = TGI_resized.flatten()
values2 = cotton_height.flatten()

# Remove NaN values from the arrays
valid_indices = np.logical_and(~np.isnan(values1), ~np.isnan(values2))
values1 = values1[valid_indices]
values2 = values2[valid_indices]

# Calculate the Pearson correlation between the two arrays
correlation = np.corrcoef(values1, values2)[0, 1]

# Print the Pearson correlation
print("Correlation:", correlation)

# Create a scatter plot
plt.scatter(values1, values2)
plt.xlabel("TGI Values")
plt.ylabel("Cotton Heights")
plt.title("Scatter Plot of TGI and Cotton Heights 1_1")

# Add the Pearson correlation line
z = np.polyfit(values1, values2, 1)
p = np.poly1d(z)
plt.plot(values1, p(values1), color='red')

# Add the Pearson correlation annotation
plt.annotate(f"Pearson correlation: {correlation:.2f}", xy=(0.6, 0.1), xycoords='axes fraction', color='red')

# Show the plot
plt.show()
```
