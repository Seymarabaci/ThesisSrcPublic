```python
import rasterio
import matplotlib.pyplot as plt
import numpy as np

# Load the red band and NIR band rasters
with rasterio.open(r'/yeni_sentinel2/B4_Soke11.tif') as red_ras:
    red = red_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B3_Soke11.tif') as green_ras:
    green = green_ras.read(1)
with rasterio.open(r'/yeni_sentinel2/B2_Soke11.tif') as blue_ras:
    blue = blue_ras.read(1)

# Calculate TGI
TGI = -0.5*((red-green)*175-(red-blue)*105)

TGI_filtered = np.where(TGI<0.000001,np.nan,TGI)


# Create a figure and plot the TGI image using the "RdYlGn" color map
fig, ax = plt.subplots()
im = ax.imshow(TGI_filtered, cmap="RdYlGn")

# Add a color bar to the figure
cbar = fig.colorbar(im, ax=ax)

# Show the figure
plt.show()
from rasterio.plot import show_hist
show_hist(source=TGI_filtered, bins=50, title='Histogram of TGI 1_1', 
          histtype='stepfilled', alpha=0.5)
```
